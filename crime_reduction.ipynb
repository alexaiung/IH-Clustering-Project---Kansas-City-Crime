# Data Transformation

## Creating columns

```python
# Data columns
crime['Reported_Datetime'] = pd.to_datetime(crime['Reported_Date'] + ' ' + crime['Reported_Time'], format='%m/%d/%Y %H:%M', errors='coerce')
crime['From_Datetime'] = pd.to_datetime(crime['From_Date'] + ' ' + crime['From_Time'], format='%m/%d/%Y %H:%M', errors='coerce')
crime['To_Datetime'] = pd.to_datetime(crime['To_Date'] + ' ' + crime['To_Time'], format='%m/%d/%Y %H:%M', errors='coerce')
crime['time_until_report'] = crime['Reported_Datetime'] - crime['From_Datetime']

# Binary columns
mask_crime = crime['Age'].isna()
crime.loc[mask_crime, 'is_minor'] = 1
crime.loc[~mask_crime, 'is_minor'] = 0

mask_weekday = crime['From_Datetime'].dt.weekday <= 3
crime.loc[mask_weekday, 'is_weekend'] = 0 
crime.loc[~mask_weekday, 'is_weekend'] = 1

mask_night = (crime['From_Datetime'].dt.time >= pd.to_datetime('18:00:00').time()) | ((crime['From_Datetime'].dt.time <= pd.to_datetime('6:00:00').time()))
crime.loc[mask_night, 'is_night'] = 1
crime.loc[~mask_night, 'is_night'] = 0

# Age columns
mask_minor = crime['Age'].isna()
mask_young_adult = (crime['Age'] >= 18) & (crime['Age'] <= 35)
mask_adult = (crime['Age'] > 35) & (crime['Age'] <= 60)
mask_elder = (crime['Age'] > 60)

crime.loc[mask_minor, 'cat_age'] = 'minor'
crime.loc[mask_young_adult, 'cat_age'] = 'young_adult'
crime.loc[mask_adult, 'cat_age'] = 'adult'
crime.loc[mask_elder, 'cat_age'] = 'mask_elder'
```

## Dropping data

```python
crime.columns
```

```python
# Deletar tudo que não é Kansas City
mask_kansas_city = crime['City'] != 'KANSAS CITY'
crime = crime.drop(crime[mask_kansas_city].index)
```

```python
crime = crime.drop(['Reported_Date', 'Reported_Time', 'To_Date', 'To_Time', 'From_Date', 'From_Time', 'Description', 'Beat', 'Address', 'City', 'Zip Code'], axis=1)
crime = crime.drop(crime[crime['Involvement'] != 'VIC'].index)
```

# Clusterizing

```python
cat_vars = ['Offense', 'Rep_Dist', 'Race', 'Sex', 'Firearm Used Flag', 'is_minor', 'is_weekend', 'cat_age', 'is_night']
ohe = OneHotEncoder(drop='first', min_frequency=0.01, handle_unknown='infrequent_if_exist')
preprocessor = ColumnTransformer(transformers=[('onehot', ohe, cat_vars)])
transformed_data = preprocessor.fit_transform(crime)
```

```python
ward_2 = AgglomerativeClustering(n_clusters=2, linkage='ward')
crime['label_kmeans'] = ward_2.fit(crime).labels_
```

```python
sns.barplot(x=crime.labels_kmeans.value_counts().index, y=crime.labels_kmeans.value_counts())
```
